**Overfit:** When the model is so well trained that all different data into the test is out of the sitations that it's able to predict and "understand". The model is good only with the train data and cases very very similar to them.

### Importing the dataset

> Yes, is the same of the used on the previous course [model validation](http://localhost:8888/notebooks/6_model_validation/model_validation.ipynb) 

```python
import pandas as pd
from datetime import datetime

uri = "https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv"
data = pd.read_csv(uri)

sold_change = {
    'yes' : 1,
    'no'  : 0
}

current_year = datetime.today().year
data['model_age'] = current_year - data.model_year 
data['km_per_year'] = data.mileage_per_year * 1.60934 
data.sold = data.sold.map(sold_change)
data = data.drop(columns = ['Unnamed: 0', 'mileage_per_year', 'model_year'], axis=1) # Axis: 1 - Column, 0 - Row
data.head()
```

### Loading the previus project

```python
def print_results(result):
    print('Raw result: ', result)
    print('\n----------------------------------------------------------------------------\nTest Score mean: %.2f%% ' % (result['test_score'].mean() * 100))
    accuracy_mean = result['test_score'].mean()
    accuracy_std = result['test_score'].std()
    print('Test accuracy gap: [%.2f%% - %.2f%%]' % (100 * (accuracy_mean - (2 * accuracy_std)), 100 * (accuracy_mean + (2 * accuracy_std))))
```

#### Modeling the data

```python
sorted_data = data.sort_values('sold')
x = sorted_data[['model_age', 'price', 'km_per_year']]
y = sorted_data['sold']
```

#### Using Dummy Classifier

```python
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_validate
import numpy as np

SEED = 15
np.random.seed(SEED)

# Baseline (default - stratified)
dummy = DummyClassifier()
result = cross_validate(dummy, x, y, cv = 10)
print_results(result)
```

#### Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(max_depth = 2)
result = cross_validate(tree_model, x, y, cv = 10)
print_results(result)
```

```python
random_model = np.random.randint(-2, 3, size = len(data))
data['random_model'] = random_model + data['model_age']
print(data['random_model'].min())
data.head()
```

#### Using SVC

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GroupKFold
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

scaler = StandardScaler()
model = SVC()
pipeline = Pipeline([('StandardScaler', scaler), ('SVC', model)])

cross_val = GroupKFold(n_splits = 10)
result = cross_validate(pipeline, x, y, cv = cross_val, groups = data.random_model)
print_results(result)
```

### Printing and analysing results

```python
def simplified_print_results(result):
    test_accuracy_mean = result['test_score'].mean()
    train_accuracy_mean = result['train_score'].mean()
    print('mean (train): %.4f%%  mean (test): %.4f%%' % (train_accuracy_mean * 100, test_accuracy_mean * 100))
```

```python
def tree_decision(max_depth = 3, print_results = False):
    SEED = 20
    np.random.seed(SEED)
    
    tree_model = DecisionTreeClassifier(max_depth = max_depth)
    result = cross_validate(tree_model, x, y, cv = 10, return_train_score = True)
    
    test_score = result['test_score'].mean()
    train_score = result['train_score'].mean()
    table = [max_depth, train_score, test_score]
    
    if(print_results):
        print('Depth: ', max_depth)
        print('mean (train): %.4f%%  mean (test): %.4f%%' % (train_score * 100, test_score * 100))
    
    return table
```

```python
max_depth = 30
results = [tree_decision(max_depth = depth) for depth in range(1, max_depth + 1)]
```

> The best choose, in this case, is `max_depth = 3`

> This is an example of *overfitting*, where the model is excelent with train data but is much worse with test data

```python
df_results = pd.DataFrame(results, columns = ['max_depth', 'train_score', 'test_score'])
df_results
```

```python
from seaborn import lineplot
import matplotlib.pyplot as plt

lineplot(x = 'max_depth', y = 'train_score', data = df_results)
lineplot(x = 'max_depth', y = 'test_score', data = df_results)
plt.title('Exemplifying a overfit', size = 8)
plt.suptitle('Test Score vs Max depth of a tree classifier')
plt.legend(['train', 'test'])
```

## 2nd Class: Exploring 2 dimensions od hyperparameters

> Date: July 7, 2020

### DecisionTreeClassifier

**Leaf:** The last blocks of a tree or a ramification.
    - Example:
     _______________                         _______________                         _______________               
    |   Block One   |                       |   Block Two   |                       |  Block Three  |             
    |_______________|                       |_______________|                       |_______________|
            |                                        |                                      |
     _______|________                        ________|_______                        _______|________             
    | Block One_two |                       | Block Two_two |                       |Block Three_two|             
    |_______________|                       |_______________|                       |_______________|
            |                                        |                                      |    
     _______|________                        ________|_______                        _______|________             
    |     Leaf 1    |                       |    Leaf 2     |                       |     Leaf 3    |             
    |_______________|                       |_______________|                       |_______________|
    

**min_samples_leaf:** Minimum of samples in the block to be used in the classification proccess


- Now, we are going to use 2 hyperparameters: min_samples_leaf and max_depth
- To define the best values, we can use the same way used with max_depth

```python
# Fix function tree_decision to receive min_samples_leaf as a parameter
# Use, show and return the value
# Run a for loop to make a dataFrame showing the score with defined hyperparameters
# Make a function to do all this things
# Sort by test score
```

```python
def tree_decision(max_depth = 3, print_results = False, min_samples_leaf = 1):
    SEED = 20
    np.random.seed(SEED)
    
    tree_model = DecisionTreeClassifier(max_depth = max_depth, min_samples_leaf = min_samples_leaf)
    result = cross_validate(tree_model, x, y, cv = 10, return_train_score = True)
    
    test_score = result['test_score'].mean()
    train_score = result['train_score'].mean()
    table = [max_depth, min_samples_leaf, (train_score * 100), (test_score * 100)]
    
    if(print_results):
        print('Depth: ', max_depth)
        print('Samples Leaf: ', min_samples_leaf)
        print('mean (train): %.4f%%  mean (test): %.4f%%' % (train_score * 100, test_score * 100))
    
    return table
```

```python
max_depth = 16
min_samples_leaf = 256
results = []

for depth in range(1, max_depth + 1):
    for samples_leaf in np.linspace(1, min_samples_leaf, 32):
       results.append(tree_decision(max_depth = depth, min_samples_leaf = int(samples_leaf))) 
```

```python
df_results = pd.DataFrame(results, columns = ['max_depth', 'min_samples_leaf' , 'train_score', 'test_score'])
df_results = df_results.sort_values(['test_score', 'min_samples_leaf'], ascending = False)
df_results.head(10)
```

> There is a big correlation between max_depth and train score

## 3rd Class: Working with 3 or more dimensions

> Date: July 7, 2020

**Split:** A Cut in the tree that genarate another ramification.
    - Example:
     _______________                         _______________                         _______________               
    |   Block One   |                       |   Block Two   |                       |  Block Three  |             
    |_______________|                       |_______________|                       |_______________|
            |____split____                            |                                      |
            |            \                           |                                      |
     _______|________     \__                ________|_______                        _______|________           
    | Block One_two |  _______|________      | Block Two_two |                       |Block Three_two| 
    |_______________| |Block One_tree |      |_______________|                       |_______________|
            |         |_______________|              |                                      |    
     _______|________                        ________|_______                        _______|________             
    |     Leaf 1    |                       |    Leaf 2     |                       |     Leaf 3    |             
    |_______________|                       |_______________|                       |_______________|
    


```python
def tree_decision(max_depth = 3, print_results = False, min_samples_leaf = 1, min_samples_split = 1):
    SEED = 20
    np.random.seed(SEED)
    
    tree_model = DecisionTreeClassifier(max_depth = max_depth, 
                                        min_samples_leaf = min_samples_leaf, 
                                        min_samples_split = min_samples_split)
    result = cross_validate(tree_model, x, y, cv = 10, return_train_score = True)
    
    test_score = result['test_score'].mean()
    train_score = result['train_score'].mean()
    
    fit_time = result['score_time'].mean()
    score_time = result['fit_time'].mean()
    
    table = [max_depth, min_samples_leaf, 
             min_samples_split, (train_score * 100),
             (test_score * 100), fit_time,
             score_time]
    
    if(print_results):
        print('Depth: ', max_depth)
        print('Samples leaf: ', min_samples_leaf)
        print('Samples split: ', min_samples_split)
        print('mean (train): %.4f%%  mean (test): %.4f%%' % (train_score * 100, test_score * 100))
    
    return table
```

```python
max_depth = 5
min_samples_leaf = np.linspace(200, 300, 10)
min_samples_split = np.linspace(10, 100, 10)
results = []

for depth in range(1, max_depth + 1):
    for samples_leaf in min_samples_leaf:
        for samples_split in min_samples_split:
           results.append(tree_decision(max_depth = depth, 
                                        min_samples_leaf = int(samples_leaf),
                                        min_samples_split = int(samples_split))) 
```

```python
df_results = pd.DataFrame(results, columns = ['max_depth', 'min_samples_leaf', 'min_samples_split', 'train_score', 'test_score', 'fit_time', 'score_time'])
df_results = df_results.sort_values(['test_score', 'min_samples_leaf', 'min_samples_split'], ascending = False)
df_results.head(10)
```

## 4th Class: Hyperparameters search with GridSearchCV 

> Date: July 9, 2020

**Library:** [Sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

GridSearchCV is a tool made to search for the best parameters into the model (like what we were doing)

```python
# Define a parameters space (dictonary)
# Run GridSearchCV (with Decision tree classifier and GroupKFold)
# Fit the model with the sorted data (x and y)
# Print the best model
# Print the accuracy of the best model
```

```python
from sklearn.model_selection import GridSearchCV

SEED = 15
np.random.seed(SEED)

param_grid = {'max_depth': range(2, 5),
                   'min_samples_leaf': range(250, 300, 5),
                   'min_samples_split': list(range(2, 100, 5)),
                   'criterion' : ['gini', 'entropy']}

grid = GridSearchCV(DecisionTreeClassifier(),
                    param_grid,
                    cv = GroupKFold(n_splits = 10))
```

```python
sorted_data = data.sort_values('sold')

x = sorted_data[['model_age', 'price', 'km_per_year']]
y = sorted_data['sold']

grid.fit(x, y, groups = data['random_model'])
```

```python
df_search = pd.DataFrame(grid.cv_results_)
df_search.sort_values('rank_test_score')
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
from sklearn.metrics import accuracy_score

results = grid.predict(x)
score = accuracy_score(results, y)
print('Test accuracy: %.2f%% ' % (score * 100))
```

## 5th Class: Nested Cross Validation and validating the chosen model

> Date: July 9, 2020

**Library:** [Sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)


### Steps

1. Do a cross validation with the created model instead of use `.predict()`, as recommended in the [documentation](https://scikit-learn.org/stable/auto_examples/model_selection/plot_nested_cross_validation_iris.html#:~:text=Nested%20cross%2Dvalidation%20(CV),also%20need%20to%20be%20optimized.&text=Model%20selection%20without%20nested%20CV,model%20and%20overfit%20the%20data.). This alternative way is named **Nested cross validation**;

2. Verify the error. Searching about the error message in Google, we can see that this is a bug (as commented and discussed [here](https://github.com/scikit-learn/scikit-learn/issues/7646)). This bug is related to the use of GroupKFold together with Nested cross validation;

3. Change the model using Kfold (with shuffle) instead of GroupKFold and try to use cross validation again;

4. Generate a graph to show how is our decision tree.


> Nested Cross validation uses the best model into the models generated by the cross validation during the training process

```python
from sklearn.model_selection import cross_val_score

cross_val = GroupKFold(n_splits = 10)
result = cross_val_score(grid, x, y, cv = cross_val, groups = data.random_model)
print_results(result)
```

> Since this is an error into the sklearn, we can't obtain the score of the results using groupKFold. Instead, we'll use the KFold with the parameter `shuffle = True`. 

```python
from sklearn.model_selection import KFold

grid = GridSearchCV(DecisionTreeClassifier(),
                    param_grid,
                    cv = KFold(n_splits = 10, shuffle = True))
grid.fit(x, y, groups = data['random_model'])
```

```python
from sklearn.model_selection import KFold

cross_val = KFold(n_splits = 10, shuffle = True)
result = cross_val_score(grid, x, y, cv = cross_val)
```

```python
def print_scores(result):
    result_mean = result.mean()
    result_std = result.std()
    print('Accuracy interval: %.4f%% - %.4f%%' % (((result_mean - result_std) * 100), ((result_mean + result_std) * 100)))
```

```python
print_scores(result)
```
