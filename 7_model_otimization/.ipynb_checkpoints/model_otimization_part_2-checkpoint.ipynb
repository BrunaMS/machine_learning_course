## Exploring randomly

### 1. Data Management

#### 1.1. Import and manage data 

```python
import numpy as np
import pandas as pd
from datetime import datetime

uri = "https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv"
data = pd.read_csv(uri)

sold_change = {
    'yes' : 1,
    'no'  : 0
}

current_year = datetime.today().year
data['model_age'] = current_year - data.model_year 
data['km_per_year'] = data.mileage_per_year * 1.60934 
data.sold = data.sold.map(sold_change)
data = data.drop(columns = ['Unnamed: 0', 'mileage_per_year', 'model_year'], axis=1) # Axis: 1 - Column, 0 - Row
random_model = np.random.randint(-2, 3, size = len(data))
data['model'] = random_model + data['model_age']
print(data['model'].min())
data.head()
```

#### 1.2. Defining data and parameters to train and test the model

```python
sorted_data = data.sort_values('sold')
x = sorted_data[['model_age', 'price', 'km_per_year']]
y = sorted_data['sold']
```

### 2. Define and fit the model

#### 2.1. Using `RandomizedSearchCV` with `n_iter`

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold
import numpy as np

SEED = 17
np.random.seed(SEED)

param_grid = {'max_depth': range(2, 5),
                   'min_samples_leaf': range(250, 300, 5),
                   'min_samples_split': list(range(2, 100, 5)),
                   'criterion' : ['gini', 'entropy']}

random_model = RandomizedSearchCV(DecisionTreeClassifier(),
                            param_grid,
                            cv = KFold(n_splits = 5, shuffle = True),
                            n_iter = 30,
                            random_state = SEED)

random_model.fit(x, y)
results = pd.DataFrame(random_model.cv_results_)
```

### 3. Print results

#### 3.1. Generate a DataFrame 

```python
results.sort_values('rank_test_score').head()
```

#### 3.2. Define the best estimator

```python
best_model = random_model.best_estimator_
best_model
```

#### 3.3.  Use `cross_val_score` to obtain the results of the best model

```python
from sklearn.model_selection import cross_val_score

best_results = cross_val_score(best_model, x, y,
                cv = KFold(shuffle = True, n_splits = 5))
```

```python
def print_scores(result):
    result_mean = result.mean()
    result_std = result.std()
    print('Mean: %.4f%% \nAccuracy interval: %.4f%% - %.4f%%' % ((result_mean * 100), ((result_mean - result_std) * 100), ((result_mean + result_std) * 100)))
```

```python
print_scores(best_results)
```

#### 3.4. Compare

The result obtained with `GridSearchCV` was **Accuracy interval: 77.6639% - 79.7161%**

This means that, using much less time and iterations, we obtained a very similar result.

#### 3.5. Plot the tree

```python
from sklearn.tree import export_graphviz
import graphviz

features = x.columns
exp_graph = export_graphviz(best_model, filled = True, 
                            rounded = True, 
                            class_names = ['Yes', 'No'],
                            feature_names = features)
graph = graphviz.Source(exp_graph)
graph
```

## 1. Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
import time 

np.random.seed(SEED)

max_depth = [i for i in range(2, 5)]
max_depth.append(None)

param_grid = {'max_depth'        : max_depth,
              'min_samples_leaf' : randint(32, 128),
              'min_samples_split': randint(32, 128),
              'criterion'        : ['gini', 'entropy'],
              'bootstrap'        : [True, False], 
              'n_estimators'     : randint(10, 100)}

time_init = time.time()
random_model = RandomizedSearchCV(RandomForestClassifier(random_state = SEED),
                            param_grid,
                            cv = KFold(n_splits = 5, shuffle = True),
                            n_iter = 30,
                            random_state = SEED)

random_model.fit(x, y)
time_end = time.time()

print("Time to define and fit the model: ", (time_end - time_init))
```

## 5th Class:  Hyperparameters optimization without cross validation

> Date: July 16, 2020

### How to optimize my model and the data used to train and without using cross validation?

The difference between a cross validation with 2 steps and the method used below is that, in a cross validation, we use half of the data to train AND test 2 times (first use one and then another). Doing what we see below, we train with a percent of the data (1 - test_size) and test with the remaining data, defined in the parameter `test_size`.

### Steps

1. Use another method (splitter class) to mix and select the train data instead of `KFold`
   
    1.1. Import `StratifiedShuffleSplit`;
    
    1.2. Test using `StratifiedShuffleSplit`; 
    
    1.3. Define `n_splits = 1` and `test_size = 0.2`;


2. Split the data in 3 parts: **train** (60%), **test** (20%) and **validation** (20%)
    
    2.1. Use `train_test_split` to do that;
    
        2.1.2. Use `shuffle = True` and `test_size = 0.20`
    
    2.2. Into `StratifiedShuffleSplit`, change test_size to 0.25 (seeing that there is only 80% of the data, to use exatly 20% of the total of samples, we need of 20/80 = 0.25).
    

### 1. Using another splitter class

```python
from sklearn.model_selection import StratifiedShuffleSplit, train_test_split 

SEED = 15
np.random.seed(SEED)

param_grid = {'max_depth'        : randint(3, 6),
              'min_samples_leaf' : randint(32, 128),
              'min_samples_split': randint(32, 128),
              'criterion'        : ['gini', 'entropy'],
              'bootstrap'        : [True, False], 
              'n_estimators'     : randint(10, 100)}

time_init = time.time()

[x_train_test, x_validation, y_train_test, y_validation] = train_test_split(x, y, 
                                                                            shuffle = True, 
                                                                            test_size = 0.2, 
                                                                            random_state = SEED)

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.25, random_state = SEED)
split_model = RandomizedSearchCV(RandomForestClassifier(random_state = SEED),
                                 param_grid,
                                 cv = split,
                                 n_iter = 5,
                                 random_state = SEED)

split_model.fit(x_train_test, y_train_test)
time_end = time.time()

print("Time to define and fit the model: %.2f sec" % (time_end - time_init))
df_results = pd.DataFrame(split_model.cv_results_)
df_results.head()
```

### 2. Print scores

```python
time_init = time.time()
results = cross_val_score(random_model, x_validation, y_validation,
                          cv = split)
time_end = time.time()
```

```python
print("Time to use a cross val score: %.2f sec" % (time_end - time_init))
print(x_train_test.shape, x_validation.shape, y_train_test.shape, y_validation.shape)
```

```python
print('Mean: %0.2f%%' % (results * 100))
```
